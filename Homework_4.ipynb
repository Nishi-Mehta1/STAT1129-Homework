```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 6, 7, 8, 9, 10])

plt.plot(x, y)

plt.show()
```

```python
x1 = np.array([3, 6, 9, 12])
y1 = np.array([2, 8, 1, 10])

plt.plot(x1, y1, marker = 'D')

plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 14, 10, 12])

plt.plot(x, y, color = 'r', linestyle ='dashed' , marker = 'D', ms = 10, mec = 'g', mfc = 'g')

plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 2, 3, 4, 5])
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([5, 6, 7, 8, 9])
x2 = np.array([0, 1, 2, 3, 4])
y2 = np.array([9, 10, 11, 12, 13])

plt.plot(x, y, x1, y1, x2, y2)

plt.show()
```

```python
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}

for i in marks:
    print(i , marks[i])
```

```python
x = np.array(['Andy', 'Amy', 'James', 'Jules', 'Arthur'])
y = np.array([88, 66, 90, 55, 77])

plt.bar(x, y)
plt.title('Marks of Students')

plt.show()
```

```python
plt.pie(y, labels = x)
plt.title('Marks of Students')

plt.legend(bbox_to_anchor = (1.5, 1.0), loc = 'upper right')

plt.show()
```

```python
from importlib import reload
plt = reload(plt)

#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([0, 1, 2, 3])
x1 = np.array([0, 1, 2, 3])
y1 = np.array([1, 2, 3, 4])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([2, 3, 4, 5])

plt.subplot(2, 3, 1)
plt.plot(x, y, x1, y1, x2, y2)
plt.title("Lines")

#plot 2
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 5, 4, 2])

plt.subplot(2, 3, 2)
plt.bar(x, y)
plt.title("Bar Graph")

#plot 3
x = np.array([30,15, 50, 5])
y = np.array(["Apple", "Orange", "Pineapple", "Blueberry"])

plt.subplot(2, 3, 3)
plt.pie(x, labels = y)
plt.title("Pie Chart")

#plot 4

x = np.array([2, 4, 6, 8])
y = np.array([2, 4, 6, 8])

plt.subplot(2, 3, 4)
plt.plot(x, y)
plt.grid(color = 'green', linestyle = '-', linewidth = 0.5)
plt.title("Grid")

#plot 5
x = np.array([4, 3, 6, 5])
y = np.array([5, 6, 6, 3])

plt.subplot(2, 3, 5)
plt.scatter(x, y)
plt.title("Scatter")

#plot 6
x = np.random.normal([10, 11, 12])

plt.subplot(2, 3, 6)
plt.hist(x)
plt.title("Histogram")

plt.show()
```
