```r
table = data.frame (
    id = c(1, 2, 3, 4, 5),
    name = c("Peter","Amy", "Ryan", "Gary", "Michelle"),
    salary = c(623.30, 515.20, 611.00, 729.00, 843.25)
)
table
```

```r
new_table = cbind(table, department = (c("IT", "finance", "marketing", "HR", "operations")))
new_table
```

```r
extraction = new_table[c(1, 3, 5), c(2, 3)]
extraction
```
